```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,101)
y = np.sin(x)
SNR_DB = 1
SN = 10**(SNR_DB/10)
SN
```

```python
y_noise_gaus = np.random.normal(loc = y,scale = 1/SN)
y_noise_rand = y + (2*np.random.rand(y.shape[0])-1)/SN
y_noise_sino = y + np.sin(100*np.pi*x)/SN
```

```python
plt.figure()
plt.plot(y)
plt.plot(y_noise_gaus,'--')
```

```python
plt.figure()
plt.plot(y)
plt.plot(y_noise_rand,'--')
```

```python
plt.figure()
plt.plot(y)
plt.plot(y_noise_sino,'--')
```
