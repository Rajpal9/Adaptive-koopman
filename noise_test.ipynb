```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,501)
y = np.sin(x)
SNR_DB = 10
SN = 100**(SNR_DB/20)
SN
```

```python
y_noise_gaus = np.random.normal(loc = y,scale = 1/SN)
y_noise_rand = y + (2*np.random.rand(y.shape[0])-1)/SN
y_noise_sino = y + np.sin(100*np.pi*x)/SN
```

```python
plt.figure()
plt.plot(y, label = 'nominal')
plt.plot(y_noise_gaus,'--', label = 'noise')
```

```python
plt.figure()
plt.plot(y)
plt.plot(y_noise_rand,'--')
```

```python
plt.figure()
plt.plot(y)
plt.plot(y_noise_sino,'--')
```
